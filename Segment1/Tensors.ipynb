# Scalars in Tensorflow
- tensors created with wrapper
- tf.Variable
- tf.constant
- tf.placeholder
- tf.SparseTensor

-- easity convert to and fromm numpy arrays

```python
import tensorflow as tf
print(tf.__version__)

```

```python
xtf = tf.Variable(23,dtype=tf.int16)
xtf
```

```python
xtf.shape
```

```python
ytf = tf.Variable(3,dtype=tf.int16)
ytf
```

```python
sumi = xtf + ytf
sumtf = tf.add(xtf,ytf)
print(sumi)
print(sumtf)
```

```python
sumtf.numpy()
sumtf
```

```python
tffloat = tf.Variable(25,dtype=tf.float16)
tffloat
```

2.Image width in pixecls e.g. 28
3. Number of color chanels e.g. Full color images

```python
images_pt = torch.rand([32,28,28,3])
```
