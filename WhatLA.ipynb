```python
import numpy as np
import  matplotlib.pyplot as plt
```

```python
t = np.linspace(0,40,1000) 
# start,finish , n points
```

```python
# distance travelled by robber
dr = 2.5*t
# distance travelled by sherrif
ds = 3 * (t-5)
```

```python
fig , ax = plt.subplots()
plt.title("A Bank Robber caught")
plt.xlabel("Time (minutes)")
plt.ylabel("Distance (miles)")
ax.set_xlim(0,40)
ax.set_ylim(0,100)
ax.plot(t,dr,label="Robber",color="red")
ax.plot(t,ds,label="Sherrif",color="blue")
plt.axvline(x=30, color='green', linestyle='--')
_ = plt.axhline(y=75, color='purple', linestyle='--')
```
